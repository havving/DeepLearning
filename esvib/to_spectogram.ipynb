```python
import matplotlib.pyplot as plt
import time as ti
```

```python
noisefile = "/src/data/RawData/normal/310-1-5_190614_0210V_cvrt.txt"
noise = []
```

```python
print("read data : %s" %noisefile)

with open(noisefile, "r") as f:
    lines = f.readlines()
    for line in lines:
        data = line.split(' ')
        noise.append(int(data[2]))
        
print("read complete.")
```

```python
targetStart = 50000
noise = noise[targetStart:targetStart * 10000]
```

```python
start = ti.time()
plt.specgram(noise, Fs=2000)
end = ti.time()

print("elapsed:%f" %(end - start))

#plt.gray()
plt.axis('off')
plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)
plt.show()
```
